```python
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import find_peaks

# Read the CSV file
df = pd.read_csv("Spectrogram_Scopy_Strings/LowE.csv")

# Find all peaks in the amplitude column
peaks, _ = find_peaks(df['Amplitude CH1(dBFS)'])

# Filter out peaks with a frequency below 50Hz
filtered_peaks = [i for i in peaks if df['Frequency(Hz)'].iloc[i] >= 80 and df['Amplitude CH1(dBFS)'].iloc[i] > -70]

# Sort the filtered peaks by amplitude value (in descending order) and select the top 4
top_peaks = sorted(filtered_peaks, key=lambda i: df['Amplitude CH1(dBFS)'].iloc[i], reverse=True)[:4]

# Print details of the top peaks
freq_peak_list = df['Frequency(Hz)'].iloc[top_peaks].values.tolist()
freq_peak_list.sort()
print("Peak frequencies:", freq_peak_list)

# Plot amplitude vs frequency
plt.figure(figsize=(8, 6))
plt.plot(df['Frequency(Hz)'], df['Amplitude CH1(dBFS)'], label="Amplitude")
# Mark the top peaks
plt.plot(df['Frequency(Hz)'].iloc[top_peaks],
         df['Amplitude CH1(dBFS)'].iloc[top_peaks],
         "x", markersize=10, label="Peaks", color='red')
plt.xlabel("Frequency (Hz)")
plt.ylabel("Amplitude")
plt.title("Amplitude vs Frequency (Low E String)")
plt.legend()
plt.grid(True)
plt.show()
```
